# Image to Pencil Sketch with Python

This project based on converting image to it's different color format. In order to convert the normal image to a sketch, we will change its original RGB values and assign its RGB values similar to grey, in this way a sketch of the input image will be generated. We are going to use severals of modules like cv2, numpy, matplotlib. we can create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just like arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Cat.jpg")
```

```python
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.title("Image",fontstyle='italic', fontsize=14,color="blue")
plt.imshow(image2)
plt.axis(False)
plt.show()
```

```python
plt.imshow(image)
plt.title("BGR image",fontstyle='italic', fontsize=14,color="blue")
plt.axis(False)
plt.show()
```

```python
G_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.title("Gray Image",fontstyle='italic', fontsize=14,color="blue")
plt.imshow(G_image)
```

```python
InvertedBlurImage = cv2.bitwise_not(G_image)
plt.title("Inverted Blurred Image",fontstyle='italic', fontsize=14,color="blue")
plt.imshow(InvertedBlurImage)
plt.show()
```

```python
blur_image = cv2.GaussianBlur(InvertedBlurImage, (111,111), 0)
plt.title("Blur Image",fontstyle='italic', fontsize=14,color="blue")
plt.imshow(blur_image)
plt.axis(False)
plt.show()
```

```python
RGBSketchImage = cv2.divide(G_image, InvertedBlurImage, scale=256.0)
plt.imshow(RGBSketchImage)
plt.title("RGB Sketch Image",fontstyle='italic', fontsize=14,color="blue")
plt.axis(False)
plt.show()
```

```python
cv2.imwrite("Cat.png", RGBSketchImage)
```

```python
plt.figure(figsize=(11,11))
PencilSketchImage = cv2.cvtColor(RGBSketchImage,cv2.COLOR_RGB2BGR)
plt.title("Pencil Sketch Image",fontstyle='italic', fontsize=20,color="blue")
plt.imshow(PencilSketchImage)
plt.axis(False)
plt.show()
```

```python
add = np.concatenate((G_image, RGBSketchImage), axis=1)
plt.figure(figsize = (10,10))
plt.title("Concatenation of two image",fontstyle='italic', fontsize=20,color="blue")
plt.imshow(add)
plt.axis(False)
plt.show()
```

```python
tri = np.concatenate((add,blur_image), axis=1)
plt.figure(figsize = (20,20))
plt.title("Concatenation of three image",fontstyle='italic', fontsize=20,color="blue")
plt.imshow(tri)
plt.axis(False)
plt.show()
```
